```python
import pandas as pd
```

```python
employee = pd.read_csv('employee.csv')
```

```python
employee = employee.drop(['EmployeeCount','EmployeeNumber', 'Over18', 'StandardHours'], axis=1)
```

```python
employee['Attrition'].unique()
```

```python
employee.loc[employee['Attrition']=='No', 'Attrition']=0
employee.loc[employee['Attrition']=='Yes', 'Attrition']=1
```

```python
employee['Attrition'].unique()
```

```python
employee.isna().sum()
```

```python
import numpy as np
employee_num =employee[employee.select_dtypes(include=[np.number]).columns.tolist()]
employee_num.info()
```

```python
employee_dummies =employee.select_dtypes(include=['object'])

employee_dummies.info()
```

```python
employee_dummies =pd.get_dummies(employee_dummies)
```

```python
employee_dummies.head()
```

```python
employee_combined = pd.concat([employee_num,employee_dummies], axis=1)
```

```python
employee_combined.info()
```

```python
employee_combined.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_y = employee_combined['Attrition']
```

```python
train_x = employee_combined.drop(['Attrition'], axis = 1)
```

```python
x_train,x_test,y_train , y_test = train_test_split(train_x, train_y, test_size = 0.3, random_state = 20)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100) # creating random forest decision tree object
```

```python
#train the model using random forest object and training data
rf.fit(x_train, y_train)
```

```python
#Predict the response for the data
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
precision = precision_score(y_test, y_pred) # measures false positives 
recall = recall_score(y_test, y_pred) # measures false negatives
roc_auc = roc_auc_score(y_test, y_pred)

print('Accuracy :', accuracy)
print('f1 :', f1)
print('precision :', precision)
print('recall :', recall)
print('roc_auc :', roc_auc)
```

```python
from time import time
from scipy.stats import randint as sp_randint
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
#utility method for reporting parameters
def report(results, ntop=3):
    for i in range(1, ntop+1):
        candidates = np.flatnonzero(results['rank_test_score'] ==i) ##making same dimension, where ranktest score is nonzero
        for candidate in candidates:
            print("Model with rank : {0}".format(i))
            print("Model validation score: {0:.3f}(std: {1:.3f})".format(results['mean_test_score'][candidate], 
                                                                         results['std_test_score'][candidate]))
            print ("Parameters : {0}".format(results['params'][candidate]))
            print("")
```

```python
#simplify parameters and distributions to sample from
para_dict = {"max_depth" : [3, None],
            "max_features": sp_randint(1,18),
            "min_samples_split": sp_randint(2,11),
            "bootstrap": [True, False],
            "criterion": ['gini', 'entropy']}
```

```python
#run randomized search
n_iter_search = 20
random_search = RandomizedSearchCV(rf, param_distributions= para_dict, n_iter = n_iter_search, cv= 5) ##cv = cross validaton k-fold = 5
```

```python
start = time()
random_search.fit(x_train, y_train)
print ("RandomizedSearchCV took %.2f seconds for %d candidates"
       "parameter settings."% ((time()-start), n_iter_search))
report (random_search.cv_results_)
```

```python
##Grid Search
# Use a full grid over all parameters
param_grid = {"max_depth" : [3, None],
            "max_features": [1,3,10],
            "min_samples_split": [2,3,10],
            "bootstrap": [True, False],
            "criterion": ['gini', 'entropy']}
```

```python
#run grid search
grid_search = GridSearchCV(rf, param_grid = param_grid, cv=5)
```

```python
start = time()
grid_search.fit(x_train, y_train)
print ("GridSearchCV took %.2f seconds for %d candidates"
       "parameter settings."% ((time()-start), len(grid_search.cv_results_['params'])))
report (grid_search.cv_results_)
```
